```python
#PUBG TASK

import random

class Env:
    def __init__(self):
        self.state = {
            "1": "Low Health",
            "2": "Take Cover",
            "3": "Enemy Close",
            "4": "Enemy Far",
            "5": "Shooting",
            "6": "Reload",
            "7": "Healing",
            "8": "Change Weapon",
            "9": "Revive Team Mate",
            "10": "Smoke",
            "11": "Frag Grenade",
            "12": "Drive",
        }
        self.agentsen = [random.choice(list(self.state.values())) for _ in range(5)]

    def Heal(self):
        self.state = "Healing"

    def EnemiesAhead(self):
        self.state = "Take Cover"

    def EnemyClose(self):
        self.state = "Enemy Close"
    
    def EnemyFar(self):
        self.state = "Enemy Far"
    
    def Shoot(self):
        self.state = "Shooting"
    
    def Reload(self):
        self.state = "Reload"
    
    def ChangeWeapon(self):
        self.state = "Change Weapon"
    
    def ReviveTeammate(self):
        self.state = "Revive Team Mate"
    
    def Smoke(self):
        self.state = "Smoke"
    
    def FragGrenade(self):
        self.state = "Frag Grenade"
    
    def Drive(self):
        self.state = "Drive"

    def getState(self):
        return self.agentsen

class agent:
    def act(self, state, dict):
        if state == "Low Health":
            return "Healing"
        elif state == "Take Cover":
            return "Taking Cover"
        elif state == "Enemy Close":
            return "Engaging Enemy"
        elif state == "Enemy Far":
            return "Taking a Defensive Position"
        elif state == "Shooting":
            return "Shooting at the Enemy"
        elif state == "Reload":
            return "Reloading Weapon"
        elif state == "Healing":
            return "Healing Team Mate"
        elif state == "Change Weapon":
            return "Changing Weapon"
        elif state == "Revive Team Mate":
            return "Reviving Team Mate"
        elif state == "Smoke":
            return "Throwing Smoke Grenade"
        elif state == "Frag Grenade":
            return "Throwing Frag Grenade"
        elif state == "Drive":
            return "Driving Vehicle"
        
            

def run_agent(agent, environment, steps):
    for step in range(steps):
        percept = random.choice(environment.getState())
        action = agent.act(state=percept, dict=environment.state)
        if percept == "Low Health":
            reward = environment.Heal()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Take Cover":
            reward = environment.EnemiesAhead()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Enemy Close":
            reward = environment.EnemyClose()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Enemy Far":
            reward = environment.EnemyFar()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Shooting":
            reward = environment.Shoot()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Reload":
            reward = environment.Reload() 
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Change Weapon":
            reward = environment.ChangeWeapon()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Revive Team Mate":
            reward = environment.ReviveTeammate()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Smoke":
            reward = environment.Smoke()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Frag Grenade":
            reward = environment.FragGrenade()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
        elif percept == "Drive":
            reward = environment.Drive()
            print(f"Step {step + 1}: Percept - {percept}, Action - {action}, Reward - {reward}")
            
            

agent = agent()
env = Env()

run_agent(agent = agent, environment = env, steps = 5)
```

```python
#Q1

import random

class SecuritySystem:
    def __init__(self):
        self.components = {
            "a": random.choice(["Safe", "Vulnerable"]),
            "b": random.choice(["Safe", "Vulnerable"]),
            "c": random.choice(["Safe", "Vulnerable"]),
            "d": random.choice(["Safe", "Vulnerable"]),
            "e": random.choice(["Safe", "Vulnerable"]),
            "f": random.choice(["Safe", "Vulnerable"]),
            "g": random.choice(["Safe", "Vulnerable"]),
            "h": random.choice(["Safe", "Vulnerable"]),
            "I": random.choice(["Safe", "Vulnerable"]),
        }

    def getState(self):
        return self.components

    def patchC(self, component):
        if self.components[component] == "Vulnerable":
            self.components[component] = "Safe"
            return f"Component {component} Solved"
        return f"Component {component} is already Safe"

class SecurityAgent:
    def scan(self, system):
        vulnerable_components = []
        for component, status in system.getState().items():
            if status == "Vulnerable":
                print(f"Component {component} found vulnerable and needs to be patched")
                vulnerable_components.append(component)
            else:
                print(f"Component {component} is Safe")
        return vulnerable_components

    def patch(self, system, vulnerable_components):
        for component in vulnerable_components:
            print(system.patchC(component))

def run_security_exercise(agent, system, steps):

    for step in range(steps):
        print(f"Step {step + 1}: Scanning system")
        vulnerable_components = agent.scan(system)
        
        if vulnerable_components:
            print("Patch Vulnerabilities:")
            agent.patch(system, vulnerable_components) 
        else:
            print("Nothing to patch")
        
        print("\nFinal Output:")
        print(system.getState()) 
        print("\n")


sec = SecuritySystem()
agent = SecurityAgent()

run_security_exercise(agent, sec, steps=1)
```

```python
import random

class BackupSystem:
    def __init__(self, numTasks):
        self.tasks = {
            f"Task {i+1}": random.choice(["Completed", "Failed"]) for i in range(numTasks)
        }

    def getStatus(self):
        return self.tasks

    def updateStatus(self, task, newStatus):
        self.tasks[task] = newStatus
        print(f"New {task} status: {newStatus}")

class BackupAgent:
    def scan_and_retry(self, backupSystem):
        failedTasks = []

        for task, status in backupSystem.getStatus().items():
            if status == "Failed":
                failedTasks.append(task)

        if failedTasks:
            for task in failedTasks:
                print(f"{task} Failed")
                backupSystem.updateStatus(task, "Completed")
        else:
            print("All task are completed")

def run_backup_management(agent, backupSystem, steps):
    print("Initial Status:")
    print(backupSystem.getStatus())  

    for step in range(steps):
        print(f"Step {step + 1}: Trying to do failed tasks")
        agent.scan_and_retry(backupSystem)

        print("Fianl Status:")
        print(backupSystem.getStatus()) 

backupSystem = BackupSystem(numTasks=5)
agent = BackupAgent()

run_backup_management(agent, backupSystem, steps=1)
```
